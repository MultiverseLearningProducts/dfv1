<p style='font-weight:bold;text-align:center;font-size:50px;line-height:90%;'> Big Data is data whose scale, distribution, diversity and or timeliness require the use of new technical architectures and analytics to enable insights that unlock new sources of business value. </p>

<div class='image' style='position:relative;width:100%;height:60%'>
    <img src="./Assets/background.PNG" >

</div>




[Top](#outline)
    
<p><b>Aim:</b> Understand the characteristics that make up 'Big Data' <p>
<hr>
<img src="./Assets/Session_1/five_v.PNG">
<hr>
<br>
<p style='font-size:40px;text-align:left;color:#1d98ff'><b>Volume</b> </p>
<br>
<div>
    <img src="./Assets/Session_1/volume.PNG" width=320 height=320> 
    <br>
    <ul style='display:inline-block'>
        <li> Relates to how much data there is </li>
        <li> Whether a dataset is considered 'big' or not is dependent on the volume </li>
        <li> Data is considered Big when it is too voluminous for traditional data storage methods </li>
        <li> Big Data platforms provide a way to store incredibly large amounts of data as well the ability to process it efficiently  </li>
    </ul> 
    
</div>
<br>
<br>
<br>
<hr>
<br>
<p style='font-size:40px;text-align:left;color:#1d98ff'><b>Velocity</b> </p>
<br>
<div>
    <img src="./Assets/Session_1/velocity.PNG" width=320 height=320> 
    <br>
    <ul style='display:inline-block'>
        <li> Refers to the speed of data generation </li>
        <li> For example, how fast data flows in from sources such as application logs, networks and social media sites </li>
        <li> Typically the flow will be massive and continuous </li>
        <li> Increasingly companies are looking to stream analytics and data so it available at the right time to make appropriate business decisions </li>
        <li> Big Data platforms offer solutions to dealing with fast flowing data in terms of storage and processing </li>
    </ul> 
    
</div>
<br>
<br>
<br>
<hr>
<br>
<p style='font-size:40px;text-align:left;color:#1d98ff'><b>Variety</b> </p>
<br>
<div>
    <img src="./Assets/Session_1/variety.PNG" width=320 height=320> 
    <br>
    <ul style='display:inline-block'>
        <li> Refers to the heterogeniety of the data sources </li>
        <li> In other words, how different the data sources are (unstructured vs structured etc) </li>
        <li> With large flows of data from several of sources it is possible that there will be a variety of data types and structures</li>
        <li> Big Data platforms can handle a variety of sources, structures and data types </li>
    </ul> 
    
</div>
<br>
<br>
<br>
<hr>
<br>
<p style='font-size:40px;text-align:left;color:#1d98ff'><b>Veracity</b> </p>
<br>
<div>
    <img src="./Assets/Session_1/quality.PNG" width=320 height=320> 
    <br>
    <ul style='display:inline-block'>
        <li> Refers to the quality of the data </li>
        <li> Checking the veracity of data sources is always challenging- <b> never trust data as given </b> </li>
        <li> Enquire about the data preparation- has it been normalised already? Has it been subject to statistical manipulation (i.e. outliers removed)? </li>
        <li> Where possible you should always ask to look at raw data </li>
        <li> Determining veracity can be challenging to an organisations policies as well as externally determining authorship/ownership </li>
    </ul> 
    
</div>
<br>
<br>
<br>
<hr>
<br>
<p style='font-size:40px;text-align:left;color:#1d98ff'><b>Value</b> </p>
<br>
<div>
    <img src="./Assets/Session_1/value.PNG" width=320 height=320> 
    <br>
    <ul style='display:inline-block'>
        <li> Refers to the business value of the data </li>
        <li> Having access to vast amounts of fast flowing data is useless unless it can leveraged into something of value </li>
        <li> How does this data add value to your business? What is the ROI?</li>
    </ul> 
    
</div>

<div class='image' style='position:relative;width:100%;height:60%'>

</div>



[Top](#outline)

<p><b> Aim:</b> Justify the use of Big Data technologies in your analysis by considering the benefits and drawbacks</p>
<hr>
<br>
<p style='font-size:40px;color:#1d98ff'><b>Advantages</b> </p>
<br>
<ul>
    <li> Ability to access and process large volumes of data quickly </li>
    <li> Analysis run on larger datasets will be more reliable, representative and accurate as taking samples from data will not be necessary </li>
    <li> Faster analysis leads to a reduction in human utilisation, meaning more projects can be undertaken or better resourcing of analysts </li>
    <li> Projects finished in shorter time frames can lead to increased reputation amongst customers and products being brought to market </li>
</ul>
<br>
<br>
<br>
<hr>
<br>
<p style='font-size:40px;color:#1d98ff'><b>Disadvantages</b> </p>
<br>
<ul>
    <li> It is expensive </li>
    <li> Big Data technologies only provide a small amount of free processing before charging and costs can rack up quickly </li>
    <li> Therefore you do not have time to play around or make mistakes- time is money </li>
    <li> Big Data still requires a degree of competency from the user, it is not something that can just be picked up by anybody </li>
    <li> The nature of Big Data means you are exporting data externally, potentially bringing in cybersecurity risks </li>
    <li> It can be challenging to integrate Big Data output into your system </li>
</ul>
<br>
<br>

<a id="vs_coding"></a>
<div class='image' style='position:relative;width:100%;height:60%'>
    <img src="./Assets/background.PNG" >
    <h2 style ='color:white; font-size: 70px; font-weight:bold; position:absolute; top:100px; left:100px; width:90%'> Platforms vs Coding Yourself </h2>
</div>


[Top](#outline)

<p><b> Aim: </b> Justify the use of platforms by comparing them to your own coding </p>
<hr>
<br>
<p style='font-size:40px;color:#1d98ff'><b>Advantages</b> </p>
<br>
<br>
<br>
<div id='wrap' style=display:table;clear:both>
    <div id=left style=float:left;width:50%>
        <h4 style=color:#1d98ff;font-size:30px> Platforms </h4>
        <br>
        <ul>
            <li> Easy to use and do not require technical knowledge or skills </li>
            <li> More agile as 'drag and drop' options allow for quicker building of functions/models/visualisations </li>
            <li> Greater agility leads to lower costs as less time is spent on development </li>
            <li> Relatively simple to edit what you have built </li>
            <li> Often tech support available </li>
        </ul>
    </div>
    <div id='right' style=float:right;width:50%>
        <h4 style=color:#2ee1a1;font-size:30px> Coding </h4>
        <br>
        <ul>
            <li> You have complete control over the whole process and can customise solour scheme and functionality in almost any way imaginable </li>
            <li> Data entry can be faster and more accurate (particularly when automated) </li>
            <li> Common coding langauges (R, Python, etc) are free to install and use </li>
            <li> You own the source code </li>
            <li> Most languages have active communities where you can ask for help </li>
        </ul>
    </div>
</div>
<br>
<br>
<br>
<hr>
<br>
<p style='font-size:40px;color:#1d98ff'><b>Disadvantages</b> </p>
<br>
<br>
<br>
<div id='wrap' style=display:table;clear:both>
    <div id=left style=float:left;width:50%>
        <h4 style=color:#1d98ff;font-size:30px> Platforms </h4>
        <br>
        <ul>
            <li> Can be a steep learning curve and training will be required for each different software you use </li>
            <li> While platforms often have a wide variety of functions, they are still limited to whatever was built in by the developer </li>
            <li> Security and reliability issues around you not having complete control over your code/data </li>
            <li> If you want to change product it can be difficult to migrate your models/visualisations </li>
            <li> Licenses can be expensive </li>
            <li> You do not own the source code </li>
        </ul>
    </div>
    <div id='right' style=float:right;width:50% >
        <h4 style=color:#2ee1a1;font-size:30px> Coding </h4>
        <br>
        <ul>
            <li> You have complete control over the whole process and can customise solour scheme and functionality in almost any way imaginable </li>
            <li> Learning to code is a daunting and long process </li>
            <li> Coding is constantly evolving and it is up to you to keep up with trends- which you may not have time to do </li>
            <li> Complicated codes take a long time to produce and require constant error checking and validation </li>
            <li> Easy to make mistakes but can be difficult to find the error </li>
            <li> Making small changes to large scripts not always simple and can lead to side effects </li>
            <li> Code is harder to understand for non-technical colleagues </li>
        </ul>
    </div>
</div>
